# Problemas de distribución de probabilidad

## Distribución binomial

En teoría de la probabilidad y estadística, la distribución binomial es la distribución de probabilidad discreta que da sólo dos resultados posibles en un experimento; por ejemplo, al lanzar una moneda, puede salir cara o cruz.

```python
from scipy.stats import binom

# Definir todas las funciones de probabilidad relacionadas con esta distribución.

def dbinom(x, size, prob = 0.5):
    """
    Calcula la estimación puntual de la distribución binomial.
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calcula el acumulado de la distribución binomial.
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calcula la función cuantil a partir de la distribución binomial.
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Genera variables aleatorias a partir de la distribución binomial.
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Distribución binomial(10, 0.2), dónde n = 10 es el número de ensayos y p = 0.2 es la probabilidad de éxito en un solo ensayo
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# TODO
from scipy.stats import norm

def dnorm(x, mean=0, sd=1):
    
    # Calcula la densidad de probabilidad (PDF) de la distribución normal. con media y desviacion estandar
    
    result = norm.pdf(x, loc=mean, scale=sd)
    return result

def pnorm(q, mean=0, sd=1):
    
    #Calcula la función de distribución acumulativa (CDF) de la distribución normal.
    
    result = norm.cdf(q, loc=mean, scale=sd)
    return result

def qnorm(p, mean=0, sd=1):
    
    #Calcula la función cuantil (PPF) de la distribución normal.
    
    result = norm.ppf(p, loc=mean, scale=sd)
    return result

def rnorm(n, mean=0, sd=1):
    
    #Genera variables aleatorias de la distribución normal.
    
    result = norm.rvs(loc=mean, scale=sd, size=n)
    return result
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Parámetros de la distribución normal
mean = 0
sd = 1
n = 2000

# Generar 2000 valores aleatorios a partir de una Normal(0, 1)
data_normal = rnorm(n, mean=mean, sd=sd)

# Valores posibles de la distribución normal
x = np.linspace(-4, 4, 100)

# Calcular PDF, CDF y percentiles
pdf_normal = dnorm(x, mean=mean, sd=sd)
cdf_normal = pnorm(x, mean=mean, sd=sd)
percentiles = np.linspace(0, 1, 100)
ppf_normal = qnorm(percentiles, mean=mean, sd=sd)

fig, axis = plt.subplots(3, 1, figsize=(7, 10))

# Plot PDF
axis[0].plot(x, pdf_normal, 'r-')
axis[0].set_title("Probability Density Function (PDF)")
axis[0].set_xlabel("Value")
axis[0].set_ylabel("Density")

# Plot CDF
axis[1].plot(x, cdf_normal, 'b-')
axis[1].set_title("Cumulative Distribution Function (CDF)")
axis[1].set_xlabel("Value")
axis[1].set_ylabel("Cumulative Probability")

# Plot PPF
axis[2].plot(percentiles, ppf_normal, 'g-')
axis[2].set_title("Percent-Point Function (PPF)")
axis[2].set_xlabel("Percentiles")
axis[2].set_ylabel("Quantiles")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
from scipy.stats import uniform

def dunif(x, a=0, b=1):
    
#    Calcula la densidad de probabilidad (PDF) de la distribución uniforme.

    result = uniform.pdf(x, loc=a, scale=b-a)
    return result

def punif(q, a=0, b=1):
   
    #Calcula la función de distribución acumulativa (CDF) de la distribución uniforme.
    
    result = uniform.cdf(q, loc=a, scale=b-a)
    return result

def qunif(p, a=0, b=1):

#    Calcula la función cuantil (PPF) de la distribución uniforme.

    result = uniform.ppf(p, loc=a, scale=b-a)
    return result

def runif(n, a=0, b=1):

#    Genera variables aleatorias de la distribución uniforme.

    result = uniform.rvs(loc=a, scale=b-a, size=n)
    return result
```

```python
np.random.seed(42)

# Parámetros de la distribución uniforme
a = 0
b = 1
n = 2000

# Generar 2000 valores aleatorios a partir de una Uniforme(0, 1)
data_uniform = runif(n, a=a, b=b)

# Valores posibles de la distribución uniforme
x = np.linspace(a, b, 100)

# Calcular PDF, CDF y percentiles
pdf_uniform = dunif(x, a=a, b=b)
cdf_uniform = punif(x, a=a, b=b)
percentiles = np.linspace(0, 1, 100)
ppf_uniform = qunif(percentiles, a=a, b=b)

fig, axis = plt.subplots(3, 1, figsize=(5, 7))

# Plot PDF
axis[0].plot(x, pdf_uniform, 'r-')
axis[0].set_title("Probability Density Function (PDF)")
axis[0].set_xlabel("Value")
axis[0].set_ylabel("Density")

# Plot CDF
axis[1].plot(x, cdf_uniform, 'b-')
axis[1].set_title("Cumulative Distribution Function (CDF)")
axis[1].set_xlabel("Value")
axis[1].set_ylabel("Cumulative Probability")

# Plot PPF
axis[2].plot(percentiles, ppf_uniform, 'g-')
axis[2].set_title("Percent-Point Function (PPF)")
axis[2].set_xlabel("Percentiles")
axis[2].set_ylabel("Quantiles")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```
